## NPTEL RUN NOTEBOOK

This notebook performs coreference resolution on the NPTEL Transcripts. However this might be also used to perform coreference resolution on other documents, provided they are in the format as the NPTEL Data

**Importing Libraries**

Importing the important libraries

```python
import spacy
import torch
import socket
import os
import glob

from algorithm import Coref
from utils import (encode_distance, BATCH_SIZE_PATH, SIZE_FP,
                               SIZE_FP_COMPRESSED, SIZE_FS, SIZE_FS_COMPRESSED,
                               SIZE_GENRE, SIZE_PAIR_IN, SIZE_SINGLE_IN,SIZE_EMBEDDING)
from dataset import (NCDataset, NCBatchSampler,
    load_embeddings_from_file, padder_collate,
    SIZE_PAIR_IN, SIZE_SINGLE_IN, SIZE_EMBEDDING)
```

```python
for file in glob.glob('./NPTEL_DATA/*_nltk.txt') :
    print(file)
    file_handle = open(file,'r')
    discourse_raw = file_handle.read()
    discourse = discourse_raw.split('\n')
    coref1 = Coref()
    discourse_coref_clusters = coref1.one_shot_coref_NEW(utterances=discourse)
    
    # create a list of sentences
    sentence_list = list()
    raw_sentences = coref1.get_utterances()
    for sentence_tok in raw_sentences :
        sentence = list()
        for token in sentence_tok :
            sentence.append([token.text])
        sentence_list.append(sentence)
        
    for cluster_key, cluster_values in discourse_coref_clusters.items() :
        
        mention_count = 0
        
        for value in cluster_values :
            
            # if it's the first mention then we add first occurance in order to differentiate between the first and second occurace of simmilar entities
            if mention_count == 0 :
                occurance = 'first occurance'
            else :
                occurance = 'next occurance'
                
            sentence_num = value.utterance_index
            
            first_token = sentence_list[sentence_num][value.start][0]
            sentence_list[sentence_num][value.start].insert(0,'{' + first_token)
            sentence_list[sentence_num][value.start].pop(1)
            
            last_token = sentence_list[sentence_num][value.end - 1][0]
            sentence_list[sentence_num][value.end - 1].insert(0,last_token + '}')
            sentence_list[sentence_num][value.end - 1].pop(1)
            
            sentence_list[sentence_num][value.end - 1].append('(' + cluster_key.text + ')' + '<' + occurance + ',' + str(cluster_key.start) + ',' + str(cluster_key.end) + '>')
            
    processed_sentence_list = list()
    for sentence in sentence_list :
        processed_sentence = list()
        for token in sentence :
            complete_token = ' '.join(token)
            processed_sentence.append(complete_token)
        complete_sentence = ' '.join(processed_sentence)
        processed_sentence_list.append(complete_sentence)

    corefered_discourse = ' '.join(processed_sentence_list)
    
    del coref1
    write_file = open('{}.COREFERED_PRE_TRAINED_DATA'.format(file),'w',encoding='utf-8')
    write_file.write(discourse_raw)
    
    write_file.write("\n\n")
    
    write_file.write(corefered_discourse)
    write_file.close()
```
